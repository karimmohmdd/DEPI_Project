```python
cols = [
    'HadHeartAttack', 'HadAngina', 'HadStroke', 'HadAsthma', 'HadSkinCancer',
    'HadCOPD', 'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis',
    'HadDiabetes', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty',
    'DifficultyConcentrating', 'DifficultyWalking', 'DifficultyDressingBathing',
    'DifficultyErrands', 'ChestScan', 'AlcoholDrinkers', 'HIVTesting',
    'FluVaxLast12', 'PneumoVaxEver'
]
```

```python
for col in cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isna().sum()
```

```python
df[cols] = df[cols].replace({'Yes': 1, 'No': 0})
```

```python
Q11 = df['HeightInMeters'].quantile(0.25)
Q31 = df['HeightInMeters'].quantile(0.75)
IQR1 = Q31 - Q11
Height_lower_bound = Q11 - 1.5 * IQR1
Height_upper_bound = Q31 + 1.5 * IQR1

print('Height Upper Bound:' , Height_upper_bound)
print('Height Lower Bound:' , Height_lower_bound)



Q12 = df['BMI'].quantile(0.25)
Q32 = df['BMI'].quantile(0.75)
IQR2 = Q32 - Q12
BMI_lower_bound = Q12 - 1.5 * IQR2
BMI_upper_bound = Q32 + 1.5 * IQR2

print('BMI Upper Bound:' , BMI_upper_bound)
print('BMI Lower Bound:' , BMI_lower_bound)




Q13 = df['WeightInKilograms'].quantile(0.25)
Q33 = df['WeightInKilograms'].quantile(0.75)
IQR3 = Q33 - Q13
Weight_lower_bound = Q13 - 1.5 * IQR3
Weight_upper_bound = Q33 + 1.5 * IQR3

print('Weight Upper Bound:' , Weight_upper_bound)
print('Weight Lower Bound:' , Weight_lower_bound)
```

```python
df[df['BMI'] > BMI_upper_bound].sort_values(by='BMI', ascending=False)
```

```python
df.describe()
```

```python
df['BMI'] = df['BMI'].fillna(df['BMI'].mean()) 
df['HeightInMeters'] = df['HeightInMeters'].fillna(df['HeightInMeters'].mean()) 
df['WeightInKilograms'] = df['WeightInKilograms'].fillna(df['WeightInKilograms'].mean()) 
```
